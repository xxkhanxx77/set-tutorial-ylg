# <a id='toc2_'></a>[Import Libraries](#toc0_)

```python
import os

from settrade_v2 import Investor
from dotenv import load_dotenv
```

# <a id='toc3_'></a>[Connect Settrade](#toc0_)

```python
load_dotenv(override=True)

SETTRADE_APP_ID = os.environ["SETTRADE_APP_ID"]
SETTRADE_APP_SECRET = os.environ["SETTRADE_APP_SECRET"]
SETTRADE_APP_CODE = os.environ["SETTRADE_APP_CODE"]
SETTRADE_BROKER_ID = os.environ["SETTRADE_BROKER_ID"]
SETTRADE_PIN = os.environ["SETTRADE_PIN"]

SETTRADE_ACCOUNT_NO = os.environ["SETTRADE_ACCOUNT_NO"]
```

```python
investor = Investor(
                app_id=SETTRADE_APP_ID,
                app_secret=SETTRADE_APP_SECRET,
                broker_id=SETTRADE_APP_CODE,
                app_code=SETTRADE_BROKER_ID,
                is_auto_queue = False)
```

```python
settrade_equity = investor.Equity(account_no=SETTRADE_ACCOUNT_NO)
```

# <a id='toc4_'></a>[Get Data](#toc0_)

```python
# แสดงข้อมูลของบัญชี
settrade_equity.get_account_info()
```

```python
# แสดงข้อมูล position หุ้นแต่ละตัว, ผลรวมของพอร์ตว่ากำไรขาดทุนเท่าไร
settrade_equity.get_portfolios()
```

```python
type(settrade_equity.get_account_info())
```

# <a id='toc5_'></a>[Place order](#toc0_)

```python
settrade_equity.place_order(pin=SETTRADE_PIN,
                            side="Buy",
                            symbol="PTT",
                            volume=100,
                            price=37)
```

```python
settrade_equity.place_order(pin=SETTRADE_PIN,
                            side="Sell",
                            symbol="PTT",
                            volume=100,
                            price=38)
```
